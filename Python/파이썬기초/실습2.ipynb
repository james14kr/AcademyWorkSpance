```python
#주어진 딕셔너리 데이터로 1~2번 문제를 푸세요
student = {
  'name' : 'hong',
  'korScore' : 0,
  'mathScore' : 0
}
```

```python
#1. 위 딕셔너리 데이터에 키보드로 입력한 값으로 국어점수 및 수학 점수를 변경하시오
inputKor = int(input('국어점수 입력 : '))
student['korScore'] = inputKor
inputMath = int(input('수학 점수 입력 : '))
student['mathScore'] = inputMath

print(student)
```

```python
#2. 위 틱셔너리 데이터에 총점(totalScore)를 추가하고, value에는 국어 및 수학 점수의 합을 저장

student['totalScore'] = int(student['korScore']) + int(student['mathScore'])

print(student)
```

```python
#3. 아래 딕셔너리 데이터가 매개변수로 주어지면 모든 점수의 합을 리턴하는 함수 선언 및 호출
scores = {
  'kor' : 70,
  'math' : 80,
  'eng' : 90
}

#함수 선언
def sumNum (scores) :
  resultNum = 0
  for e in scores : 
    resultNum += scores[e]
  return resultNum
```

```python
#3번 문제의 함수를 호출

sumNum(scores)
```

```python
'''
매개변수로 전달된 딕셔너리에 주어진 key와 value를 추가 한 딕셔너리를 리턴하는 함수를 선언 및 호출
첫 번째 매개변수 : 딕셔너리
두 번째 매개변수 : 추가할 key
세 번째 매개변수 : 추가할 value
'''

#4번 함수 선언

def getItem (dic, key, value) : 
  dic[key] = value
  return dic
```

```python
#4번 함수 호출

getItem(scores, 'science', 100)
```
