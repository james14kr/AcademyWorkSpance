```python
num1 = 5

#파이썬 조건문의 실행 영역은 들여쓰기로 범위를 구분한다
if num1 > 1 : 
  print('num1은 3보다 큽니다.')
else : 
  print('num1은 3보다 작습니다.')
  
```

```python
num2 = 4

#num2 변수의 값이 3보다 크면 1, 3보다 작으면 2, 3이면 0을 출력
if num2 > 3 :
  print(1)
elif num2 < 2 :
  print(2)
else : 
  print(0)
```

```python
# input() : 키보드로 데이터 입력받기
# input으로 받는 데이터는 문자열로 받음
intputedData = input('정수 입력 : ')
print(intputedData)
```

```python
#키보드로 입력받은 정수가 짝수인지 홀수인지 판별 및 출력
inputdata = int(input('정수 입력 : '))
if inputdata % 2 == 0 : 
  print("짝수입니다.")
else : 
  print("홀수입니다.")
```
