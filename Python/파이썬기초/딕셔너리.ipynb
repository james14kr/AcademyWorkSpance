```python
#Dictionary(딕셔너리) : 자바스크립트의 객체

#딕셔너리 데이터 생성
dic1 = {}
dic2 = {
  '학번' : 3,
  '이름' : 'kim',
  '점수' : 80
}

print(dic2)
print(dic2['학번'])
```

```python
#데이터 추가 및 변경
phone = {
  'modelName' : 's20',
  'price' : 1000
}

#데이터 추가
phone['color'] = 'red'

print(phone)

#데이터 수정
phone['price'] = 2000

print(phone)
```

```python
#데이터 삭제
del phone['color']
print(phone)
```

```python
#연습
family = {
  '이름' : '홍길동',
  '나이' : 20,
  '취미' : [
    '운동', '독서', '등산'
  ],
  '가족관계' : [
    {
      '관계' : '부',
      '이름' : '홍아빠'
    },
    {
      '관계' : '모',
      '이름' : '홍엄마'
    }
  ]
}

#모든 취미 출력
print(family['취미'])
#취미 갯수 출력
print(len (family['취미']))
#마지막 취미 출력
print(family['취미'][-1])
#아버지의 이름 출력
print(family['가족관계'][0]['이름'])
#어머니의 정보에 나이 추가
family['가족관계'][1]['나이'] = 60

print(family)
```
