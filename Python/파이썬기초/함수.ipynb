```python
#파이썬 함수 선언
def sayHello() :
  print('hello')

sayHello()
```

```python
#리턴이 있는 함수
def getSum() :
  return 10 + 20

getSum()
```

```python
#매개변수가 있는 함수
def getMulti(a, b) : 
  return a * b
 
getMulti(2 ,10)
```

```python
#파이썬은 리턴 데이터를 여러개 가질 수 있음
def getNumber():
  return 10, 20

a,b = getNumber()

print(a, b)
```

```python
#파이썬은 배개변수 사용 시 매개변수 이름을 자주 사용 
def tellMe(name, age, addr) : 
  print(f'이름은 {name}, 나이는 {age}, 주소는 {addr}')

tellMe('kim', 20, 'uslan')
```

```python
def tellMe2(name='미정', age=1, addr='알수없음') : 
  print(f'이름은 {name}, 나이는 {age}, 주소는 {addr}')

tellMe2()
tellMe2('aaa')
tellMe2(addr='울산')
```
