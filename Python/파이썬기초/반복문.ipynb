```python
#파이썬은 for-each문만 있음

list1 = [1, 2, 3, 4, 5, 6]
for e in list1 : 
  print(e)
```

```python
#for-each 문을 일반 for문처럼 사용하기
for e in range(0, 10) : 
  print(e)

for e in range(2, 5) : 
  print(e)

for e in range(10) : 
  print(e)
```

```python
#반복문의 index 정보 알아내기
#반복의 index를 알고 싶다면 enumeeate() 함수 매개변수에 리스트를 넣어 사용
list1 = [2, 4, 6, 8, 10]
for i, e in enumerate(list1) : 
  print(f'데이터 : {e}, index : {i}')
```
