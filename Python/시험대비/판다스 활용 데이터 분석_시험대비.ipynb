### 판다스 활용 데이터 분석

1. 키보드로 세 수를 입력받고,입력받은 세 수를 리스트에 저장하시오. 그 후 리스트에 저장된 수 중 가장 큰 수와 가장 작은 수의 차를 출력하시오.

```python
#1번 문제 풀이 작성
list1 = []
for e in range(3) :
  inputNum = int(input(f'{e+1}번째 정수 입력 : '))
  list1.append(inputNum)
print(max(list1) - min(list1))
```

2. 매개변수로 두 개의 리스트를 받아, 두 개의 리스트에 저장된 모든 데이터의 합을 리턴하는 함수를 작성하시오.  
EX> 두 매개변수 : [1,2,3] , [4,5,6] / 결과 : 21

```python
#2번 문제 풀이 작성
list2 = [1, 2, 3]
list3 = [4, 5, 6]
print(sum(list2) + sum(list3))
```

3. 매개변수로 두 개의 리스트를 받아, 두 개의 리스트에 중복으로 저장된 데이터만을 갖는  새로운 리스트를 리턴하는 함수를 작성하시오.  
EX> 두 매개변수 : [1,3,5,7,9] , [1,2,3,4,5] / 결과 : [1,3,5]

```python
#3번 문제 풀이 작성
list4 = [1, 3, 5, 7, 9]
list5 = [1, 2, 3, 4, 5]

resultList = []

for e in list4 : 
  for f in list5 : 
    if(e == f) : 
      resultList.append(e)

print(resultList)
```

4. 이름, 나이, 국어점수, 영어점수, 국어점수를 데이터로 갖는 student 딕셔너리를 만드시오. 각 데이터는 키보드로 입력받아 저장하시오.

```python
#4번 문제 풀이 작성
student = {
  '이름' : '',
  '나이' : '',
  '국어점수' : 0,
  '영어점수' : 0,
  '수학점수' : 0,
}

for e in student : 
  inputInfo = input(f'{e} 입력 : ')
  student[e] = inputInfo

print(student)
```

6. 아래 딕셔너리 데이터에 stu_grade라는 키를 추가하고, 해당 키의  데이터로  점수(stu_score)에 따른 등급 데이터를 저장하시오.  
90 <= 점수      : A등급  
80 <= 점수 < 90 : B등급  
70 <= 점수 < 80 : C등급  
      점수 < 70 : D등급 

```python
class_room_data = {
    'stu_no' : [1, 2, 3, 4, 5],
    'stu_name' : ['김자바', '이자바', '박자바', '최자바', '정자바'],
    'stu_score' : [85, 65, 70, 80, 95]
}

#6번 문제 풀이 작성

class_room_data['stu-grade'] = []

for stuNum in class_room_data['stu_no'] : 
    score = class_room_data['stu_score'][stuNum-1]
    if score >= 90 :
        class_room_data['stu-grade'].append('A')
    elif score >= 80 :
        class_room_data['stu-grade'].append('B')
    elif score >= 70 : 
        class_room_data['stu-grade'].append('C')
    else : 
        class_room_data['stu-grade'].append('D')

print(class_room_data)
```
