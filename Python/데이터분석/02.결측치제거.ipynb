```python
import pandas as pd
```

```python
df = pd.read_excel('test2.xlsx')
df
```

```python
#결측치 : 데이터 세트에서 누락된 값(비어있는 데이터, NaN 데이터)
#결측치 제거 : 비어있거나 NaN 데이터를 삭제 => 비어있거나 NaN 데이터를 특정 데이터로 채워줌
```

```python
#실습을 위해 df에 결측치를 추가
import numpy as np

df['학급'] = np.nan
df
```

```python
#결측치를 특정 데이터로 채우기

df1 = df.copy()

#fillna 함수는 nan 데이터에 특정 데이터를 채운 결과를 리턴해줌 (원본 데이터는 변경사항 없음)
result_df = df1.fillna('임시데이터')

#결측치가 채워진 데이터프레임
result_df

df1
```

```python
#특정 컬럼의 결측치 제거
df1 = df.copy()

#취미 컬럼의 결측치만 특정 데이터로 채우기
df1['취미'].fillna('없음')
```
