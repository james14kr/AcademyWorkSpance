```python
#pandas 라이브러리 사용을 위해 import
#pandas 파일을 불러와서 앞으로 'pd'라는 이름으로 사용하겠다
import pandas as pd
```

```python
#pandas 라이브러리를 사용하면 엑셀, csv, josn 딕셔너리 데이터를 표 형태의 데이터로 가공

#딕셔너리를 데이터로 DataFrame(표 형태) 만들기
data = {
  '이름' : ['김자바', '이자바', '박자바', '최캐드', '정캐드'],
  '학급' : ['자바반', '자바반', '자바반', '캐드반', '캐드반'],
  '국어점수' : [90, 80, 70, 100, 90],
  '영어점수' : [70, 80, 80, 70, 90],
  '평균' : [80.0, 80.0, 75.0, 85.0, 90.0],
  '취미' : ['독서', '', '', '요가', '축구']
}

df = pd.DataFrame(data)
```

```python
#엑셀 파일 저장을 위해 라이브러리 설치
!pip install openpyxl
```

```python
#위 df 데이터를 엘셀로 저장하기
#엑셀 파일로 저장하면 df에 있던 index 정보까지 엑셀에 저장됨
df.to_excel('test1.xlsx')

#엑셀 파일 저장 시 index를 제외하고 저장
#참 = True, 거짓 = False
df.to_excel('test2.xlsx', index=False)
```
